```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("fivethirtyeight")
```

```python
data = pd.read_csv('../data/processed/regression_ca.csv')
data
```

```python
fig = plt.figure(figsize=(12, 10))

# sns.set(rc={'axes.facecolor':'white', 'figure.facecolor':'white'})
# sns.set_style('white')

for lower,upper,y in zip(data['lower'],data['upper'],range(len(data))):
    plt.plot((lower,upper),(y,y),'-',color=sns.color_palette("Paired")[1], alpha=0.3)
    
g = sns.scatterplot(x=data['Coef'], y=data['Variable'], color=sns.color_palette("Paired")[1], s=60)
g.grid(False)
g = plt.xlabel('Coefficients', fontsize=30)
g = plt.ylabel('Variables', fontsize=30)
g = plt.xticks(fontsize=15)
g = plt.yticks(fontsize=15)
plt.hlines(y=len(data), xmin=-1, xmax=0.5, color='black')
plt.vlines(x=0, ymin=-0.5, ymax=len(data), color='black')
plt.savefig('../images/regression_ca.png', dpi=300, bbox_inches='tight')
```
