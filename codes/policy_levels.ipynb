```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("../data/OxCGRT_latest.csv")
data.head()
```

```python
countries = ['United States', 'Canada', 'China', 'Mexico', 'Japan']

data_sub = data[data['CountryName'].isin(countries)].copy()
```

```python
data_sub.fillna(0, inplace=True)
data_sub['Date'] = pd.to_datetime(data_sub['Date'], format="%Y%m%d")
data_sub['month'] = data_sub['Date'].apply(lambda x: str(x)[:4]+'-'+str(x)[5:7])
```

```python
agg_data = data_sub.groupby(['CountryName', 'month', 'Jurisdiction']).sum().reset_index()

poli = ['C1_School closing', 'C2_Workplace closing', 'C3_Cancel public events', 
        'C4_Restrictions on gatherings', 'C5_Close public transport', 'C6_Stay at home requirements',
       'C7_Restrictions on internal movement', 'C8_International travel controls']

for pol in poli:
    agg_data[pol] = agg_data[pol].apply(lambda x: 1 if x > 0 else 0)
```

```python
agg_data_nat = agg_data[agg_data['Jurisdiction']=='NAT_TOTAL'].copy()
agg_data_state = agg_data[agg_data['Jurisdiction']=='STATE_TOTAL'].copy()
```
